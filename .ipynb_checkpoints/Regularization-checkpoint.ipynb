```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm

from sklearn.linear_model import LassoCV
from sklearn.linear_model import RidgeCV
```

```python
#Install this library using the command "pip3 install ucimlrepo" in terminal
#What these variables mean can be found in the documentation here: https://archive.ics.uci.edu/dataset/45/heart+disease
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
heart_disease = fetch_ucirepo(id=45)

type(heart_disease)

X = heart_disease.data.features 
y = heart_disease.data.targets 

df_merge=pd.concat([X,y],axis=1)

df_merge.dropna(inplace=True)
```

```python
df_merge.columns
```

```python
df_merge=df_merge[["age","chol","trestbps","thalach"]]
```

```python
df_merge
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_merge.iloc[:,:-1], df_merge.iloc[:,-1], test_size=0.2, random_state=42)

# Add constant term for intercept
X_train_const = sm.add_constant(X_train)
X_test_const = sm.add_constant(X_test)
```

```python
X_train.shape,y_train.shape
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Fit LassoCV model
lasso_cv = LassoCV(cv=5, random_state=42)  # 5-fold cross-validation
lasso_cv.fit(X_train_scaled, y_train)

# Best alpha
print("Best alpha (α):", lasso_cv.alpha_)

# Predict on test set
y_pred = lasso_cv.predict(X_test_scaled)

# Evaluate performance
print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test R²:", r2_score(y_test, y_pred))
```

```python
lasso = Lasso(alpha=lasso_cv.alpha_)
lasso.fit(X_train_scaled, y_train)

y_pred = lasso.predict(X_test_scaled)

# Evaluate performance
print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test R²:", r2_score(y_test, y_pred))
```

```python
ols = LinearRegression();
ols.fit(X_train_scaled, y_train)
ols.coef_

y_pred = ols.predict(X_test_scaled)

print("Test MSE:", mean_squared_error(y_test, y_pred))

```

```python
# Fit LassoCV model
ridge_cv = RidgeCV(cv=5)  # 5-fold cross-validation
ridge_cv.fit(X_train_scaled, y_train)

# Best alpha
print("Best alpha (α):", ridge_cv.alpha_)

# Predict on test set
y_pred = ridge_cv.predict(X_test_scaled)

# Evaluate performance
print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test R²:", r2_score(y_test, y_pred))
```

```python
ridge = Ridge(alpha=ridge_cv.alpha_)
ridge.fit(X_train_scaled, y_train);

y_pred = ridge.predict(X_test_scaled)

# Evaluate performance
print("Test MSE:", mean_squared_error(y_test, y_pred))
print("Test R²:", r2_score(y_test, y_pred))
```
